```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for csv in os.listdir("./"):
    if csv.endswith(".csv"):
        df = pd.read_csv("./" + csv)
        print(csv.split(".")[0].split("_")[-1])
        print("--------------------------")
        print(df.iloc[np.argmax(df["auc"])])
        plt.figure(figsize=(4,4))
        plt.title("Grid search result " + csv.split(".")[0].split("_")[-1] + "-task")
        sns.swarmplot(df.auc, size = 4 ,color="k", label="grid search results")
        sns.violinplot(df.auc)
        plt.scatter(0,df.auc.max(), color="r", label="best grid search results")
        plt.ylabel("AUROC")
        plt.xticks([0],[" "])
        plt.legend(loc="lower right")
        plt.savefig("violin" + csv.split(".")[0].split("_")[-1] + ".eps", dpi=300, bbox_inches="tight")
        plt.show()
        print("--------------------------")
```
